```python
import os, sqlite3
```

```python
conn = sqlite3.connect('./cmudict.db')
cur = conn.cursor()
```

```python
cur.execute('''SELECT
                    word,
                    pronounciation,
                    rhyme_sound
                FROM
                    pronounciations
                INNER JOIN words ON words.word_id = pronounciations.word_id
                INNER JOIN rhyme_sounds ON rhyme_sounds.rhyme_sound_id = pronounciations.rhyme_sound_id
                WHERE
                    rhyme_sound = 'OW1'
''')
cur.fetchall()
```
